```python
import pandas as pd
def hierarchical_clustering(input_dataset, parameters, var):
    # Create a copy of the input_data for later usage
    output = input_dataset.copy()


    import numpy as np
    import plotly.figure_factory as ff
    from sklearn import preprocessing
    from sklearn.cluster import AgglomerativeClustering
    from sklearn.preprocessing import normalize
    from sklearn import metrics
    from scipy import stats
    import plotly.io as pio
    import os
    import time
    import warnings
    warnings.filterwarnings("ignore")

    start_time = time.time()

    # Collecting the output summary for flyout
    flyout_result = {'Execution Time': None}

    file_name = 'image_hierarchical_cluster_plotly.json'

    # Input Parameters
    link = parameters.get('linkage', 'ward')
    ncluster = parameters.get('ncluster', 3)

    # Input Parameter Checks
    if type(ncluster) != int:
        raise Exception("Please enter a number for the Number of Clusters.")

    if input_dataset.isnull().any().any():
        raise Exception("Please select non-null columns.")

    # To convert categorical columns to numeric format
    le = preprocessing.LabelEncoder()
    for x in range(0, output.shape[1]):
        if output.iloc[:, x].dtype == 'O':
            output.iloc[:, x] = le.fit_transform(output.iloc[:, x])
        else:
            continue

    # Normalize data in order to scale the values to a standard range
    data_scaled = normalize(output)

    # Convert the scaled data into a DataFrame
    data_scaled = pd.DataFrame(data_scaled, columns=output.columns)

    # Initialize and Fit model variable
    model = AgglomerativeClustering(n_clusters=ncluster, affinity='euclidean', linkage=link)
    print("Modelling Training Starts")
    start_time = time.time()
    predictions = model.fit_predict(data_scaled)
    end_time = time.time()
    print("Modelling Training Stops","Printing Time took for training",end_time-start_time)
    # Add the Cluster_Number generated for all rows to a column
    input_dataset['Cluster_Number'] = predictions
    data_scaled['Cluster_Number'] = predictions

    if data_scaled['Cluster_Number'].nunique() == 1:
        raise ValueError("Hierarchical Clustering is not applicable on the current data selection.")

    class_col = parameters.get('ClassColumn', None)

    # Create Dendrogram
    fig = ff.create_dendrogram(data_scaled, color_threshold=1.5, orientation='left')

    # Update Dendrogram
    fig.update_layout(width=1500, height=600,
                      title='Dendrogram for Hierarchical Clustering', title_x=0.5,
                      xaxis=dict(
                          tickangle=-45,
                          automargin=True),
                      showlegend=False
                      )

    # Update text for y_axis ticks: Row ID + Class Column
    tick_text = fig['layout']['yaxis']['ticktext']
    if class_col is not None:
        class_col = input_dataset[class_col]
        for i, val in enumerate(tick_text):
            species_val = class_col[i]
            tick_text[i] = f"{val}, {species_val}"

            # Update dendogram
            fig.update_xaxes(title_text='Distance')
            fig.update_yaxes(title_text='Index, Values', ticktext=tick_text)
    else:
        fig.update_xaxes(title_text='Distance')
        fig.update_yaxes(title_text='Index')

    pio.write_json(fig, file=os.path.join(var["outputFileDirectory"], file_name), pretty=True)
    # fig.show()

    # Calculating Clustering Scores
    silhouette_score = round(metrics.silhouette_score(output, model.labels_, metric='euclidean'), 2)
    ch_score = round(metrics.calinski_harabasz_score(output, model.labels_), 2)
    db_score = round(metrics.davies_bouldin_score(output, model.labels_), 2)

    # Creating Evaluation Metric DataFrame
    result_dic = {'Metric': ["Silhouette Score", "Calinski-Harabasz Index",
                             "Davies-Bouldin Index"],
                  'Score': [silhouette_score, ch_score, db_score]}
    result_out = pd.DataFrame.from_dict(result_dic)

    # Creating the Final Partitions DataFrame
    final_partitions = pd.DataFrame(
        columns=['Cluster', 'Frequency', 'WithinCluster_Sum_of_Squares', 'Average_Distance', 'Max_Distance'])

    for cluster in data_scaled['Cluster_Number'].unique():
        # Get rows belonging to this cluster
        cluster_data = data_scaled[data_scaled['Cluster_Number'] == cluster].iloc[:, :-1]  # Exclude the cluster number column
        # Compute the frequency
        frequency = len(cluster_data)
        # Compute the within-cluster sum of squares
        within_cluster_sum_of_squares = ((cluster_data - data_scaled.loc[cluster]) ** 2).sum().sum()
        # Compute distances to the centroid
        distances = np.sqrt(((cluster_data - data_scaled.loc[cluster]) ** 2).sum(axis=1))
        # Compute average distance and max distance
        average_distance = distances.mean()
        max_distance = distances.max()
        # Append to the Final Partitions DataFrame
        final_partitions = final_partitions.append({'Cluster': cluster,
                                                    'Frequency': frequency,
                                                    'WithinCluster_Sum_of_Squares': within_cluster_sum_of_squares,
                                                    'Average_Distance': average_distance,
                                                    'Max_Distance': max_distance}, ignore_index=True)
    final_partitions = final_partitions.sort_values(by=['Cluster'])

    # Creating Frequency of Cluster Distribution DataFrame
    cluster_point = [f"Cluster_{i}" for i in range(input_dataset['Cluster_Number'].nunique())]
    frequency_dist_df = pd.DataFrame(input_dataset['Cluster_Number'].value_counts()).sort_index()
    frequency_dist_df.index = cluster_point
    frequency_dist_df.columns = ['Cluster Frequency']

    cat_fea = input_dataset.select_dtypes(include=["object"]).columns.to_list()
    num_fea = input_dataset.select_dtypes(include=["int64", "float64"]).columns.to_list()
    cat_fea.append('Cluster_Number')

    if len(cat_fea) > 1 and len(num_fea) > 1:
        df_cat = input_dataset[cat_fea]
        cat = df_cat.groupby('Cluster_Number').agg(lambda row: stats.mode(row)[0])
        num = input_dataset.groupby(['Cluster_Number']).mean()
        cat['Cluster_Number'] = cluster_point
        num['Cluster_Number'] = cluster_point

        cluster_center = pd.merge(cat.reset_index(drop=True), num.reset_index(drop=True), on='Cluster_Number',
                                  how='left')
        cluster_center = cluster_center.set_index('Cluster_Number').reset_index()

    elif len(cat_fea) > 1:
        df_cat = input_dataset[cat_fea]
        cat = df_cat.groupby('Cluster_Number').agg(lambda row: stats.mode(row)[0])
        cluster_center = cat.reset_index()

    else:
        cluster_center = input_dataset.groupby(['Cluster_Number']).mean()
        cluster_center = cluster_center.reset_index()

    # Computing the time for execution as a table
    ex_time = round((time.time() - start_time), 4)
    if ex_time > 60:
        ex_minutes = int(ex_time / 60)
        ex_sec = round(ex_time % 60, 4)
        ex_time_refined = str(ex_minutes) + ' minutes ' + str(ex_sec) + ' sec'
    else:
        ex_time_refined = str(ex_time) + ' sec'

    # Creating the Input Parameter Selection DataFrame
    flyout_result['Execution Time'] = ex_time_refined
    flyout_result['Class Column'] = parameters.get('ClassColumn')
    flyout_result['Linkage'] = link
    flyout_result['Number of Clusters'] = ncluster
    function_summary = pd.DataFrame.from_dict(flyout_result, orient='index', columns=['Value'])

    result_out.index += 1
    cluster_center.index += 1
    flyout_dict = {
        'Input Parameter Selection': function_summary,
        'Final Partitions': final_partitions,
        'Evaluation Metrics': result_out,
        'Frequency of Cluster Distribution': frequency_dist_df,
        'Cluster Centers': cluster_center
    }

    return {"TEXT": flyout_dict, "CSV": input_dataset}


print("ey")
hierarchical_clustering(data,{},{"outputFileDirectory":"./"})
```

```python
def optimal_number_cluster(input_data, param, var):
    from scipy.spatial.distance import cdist
    import numpy as np
    from sklearn.cluster import KMeans, MiniBatchKMeans, Birch
    from sklearn_extra.cluster import KMedoids
    from sklearn.preprocessing import LabelEncoder
    from yellowbrick.cluster import KElbowVisualizer
    import matplotlib.pyplot as plt
    import os
    import pandas as pd
    import time
    start_time = time.time()

    max_iter = param.get('max_iter', 300)
    elbow_model = param.get('elbow_model', 'MiniBatchKMeans')

    # Collecting the output summary for flyout
    flyout_result = {'Execution Time': None, 'Maximum Iterations': max_iter}

    data = input_data.copy()

    le = LabelEncoder()
    for x in range(0, data.shape[1]):
        if data.iloc[:, x].dtype == 'O':
            data.iloc[:, x] = le.fit_transform(data.iloc[:, x])
        else:
            continue

    nrefs = 3
    maxClusters = 30
    gaps = np.zeros((len(range(1, maxClusters)),))
    resultsdf = pd.DataFrame({'clusterCount': [], 'gap': []})
    
    mi=0
    for gap_index, k in enumerate(range(1, maxClusters)):
        refDisps = np.zeros(nrefs)  # Holder for reference dispersion results
#         print(refDisps)
#         for i in range(nrefs):
# #             randomReference = np.random.random_sample(size=data.shape)  # Create new random reference set
# #             display(pd.DataFrame(randomReference))
# #             km = KMeans(k, max_iter=max_iter)
# #             print(km)
# #             km.fit(randomReference)
# #             refDisps[i] = km.inertia_
# #         print(refDisps)
#             mi+=1
#             print(mi)
            
        

        km = KMeans(k, max_iter=max_iter)  # Fit cluster to original data and create dispersion
        km.fit(data)
        origDisp = km.inertia_
        distort = (sum(np.min(cdist(data, km.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0])
        gap = np.log(np.mean(refDisps)) - np.log(origDisp)  # Calculate gap statistic
        gaps[gap_index] = gap  # Assign this loop's gap statistic to gaps

        output = pd.DataFrame({'clusterCount': k, 'gap': gap, 'inertia': origDisp, 'distortion': distort}, index=[0])
        resultsdf = pd.concat([output, resultsdf], ignore_index=True)
#     return km
        
    model = None
    metric = param.get('metric', 'distortion')
    if elbow_model == 'MiniBatchKMeans':
        model = MiniBatchKMeans()
    elif elbow_model == 'kmeans':
        model = KMeans()
    elif elbow_model == 'Kmedoids':
        model = KMedoids()
    elif elbow_model == 'Birch':
        model = Birch()
    visualizer = KElbowVisualizer(model, k=(2, 30), metric=metric, timings=True)
    visualizer.fit(data)  # Fit the data to the visualizer
    visualizer.finalize()
    plt.savefig(os.path.join(var["outputFileDirectory"], "opti_plot.png"))
    # visualizer.show()

    # Computing the time for execution as a table
    ex_time = round((time.time() - start_time), 4)
    if ex_time > 60:
        ex_minutes = int(ex_time / 60)
        ex_sec = round(ex_time % 60, 4)
        ex_time_refined = str(ex_minutes) + ' minutes ' + str(ex_sec) + ' sec'
    else:
        ex_time_refined = str(ex_time) + ' sec'

    flyout_result['Execution Time'] = ex_time_refined
    flyout_result['K'] = visualizer.elbow_value_
    flyout_result['Score'] = round(visualizer.elbow_score_, 4)

    # Converting the summary into a table to be displayed in the flyout
    function_summary = pd.DataFrame.from_dict(flyout_result, orient='index', columns=['Value'])

    return {'TEXT': function_summary}
```

```python
def bsas_clustering(input_dataset, parameters, var):
    """
        Performs BSAS clustering on the input dataset.

        Args:
            input_dataset (pd.DataFrame): The input dataset.
            parameters (dict): The clustering parameters.
            var (str): The variable to be clustered.

        Returns:
            dict: The flyout dictionary and the input dataset.

        Variables:

            file_name: The file name for the scatter plot.
            file_name1: The file name for the scatter plot when there are less than 3 columns.
            cluster_: The number of clusters.
            thres: The threshold value.
            model: The BSAS clustering model.
            label: The cluster labels.
            flyout_dict: The flyout dictionary.
            ex_time_refined: The refined execution time.
            flyout_dict_new: The flyout dictionary with the model summary dataframe inserted.

        """
    output = input_dataset.copy()

    # import the libraries
    from sklearn import preprocessing
    from pyclustering.cluster.bsas import bsas
    from ath.machine_learning.clustering_score import clustering_score
    from sklearn.decomposition import PCA
    import numpy as np
    import pandas as pd
    from ath.utilities.scatter_plot_clustering import scatter_plot_clustering_more_than_3cols, \
        scatter_plot_clustering_less_than_3cols
    from ath.utilities.final_partition_clustering import final_partition
    import time
    import warnings
    warnings.filterwarnings("ignore")

    start_time = time.time()

    file_name = 'image_bsas_plotly.json'
    file_name1 = 'image_bsas1_plotly.json'

    # Collecting the output summary for flyout
    flyout_result = {'Execution Time': None}

    # fetch hyperparameters
    cluster_ = parameters.get('cluster', 3)
    thres = parameters.get('thres', 1)
    model = None

    if type(cluster_) == str:
        raise Exception("Please enter numeric value for Cluster")
    if type(thres) == str:
        raise Exception("Please enter numeric value for Threshold")
    if thres < 0 or thres > 1:
        raise Exception("Please enter Threshold value between 0 and 1")
    def model_bsas(output, model):
        # Initialize and fit model variable
        model = bsas(output.values.tolist(), maximum_clusters=cluster_, threshold=thres)
        model.process()
        clusters = model.get_clusters()

        output['Cluster_Number'] = ''
        input_dataset['Cluster_Number'] = ''
        cluster = 1
        for elm in clusters:
            for mem in elm:
                output['Cluster_Number'].iloc[mem] = cluster
                input_dataset['Cluster_Number'].iloc[mem] = cluster
            cluster += 1

        if output['Cluster_Number'].nunique() == 1:
            raise ValueError("BSAS Clustering is not applicable on current data selection.")

    if (output.dtypes[output.dtypes == 'O']).any():
        # Converting categorical columns to ordinal
        le = preprocessing.LabelEncoder()
        for x in range(0, output.shape[1]):
            if output.iloc[:, x].dtype == 'O':
                output.iloc[:, x] = le.fit_transform(output.iloc[:, x])
            else:
                continue

    if len(output.columns) < 3:
        model_bsas(output, model)
        df_feat = output.drop('Cluster_Number', axis=1)
        input_dataset['Cluster_Number'] = output['Cluster_Number']
        # score_metrics = score(model_(output, model), output)
        flyout_dict = clustering_score(output, input_dataset, 'BSAS Clustering')
        # predict the labels of clusters.
        label = output['Cluster_Number']
        # Getting unique labels
        u_labels = np.unique(label)
        title = 'BSAS Clustering'
        scatter_plot_clustering_less_than_3cols(output, input_dataset, file_name1, var, title)
    else:
        # fitting the kmeans model
        model_bsas(output, model)
        df_feat = output.drop('Cluster_Number', axis=1)
        # Principal Component Analysis
        pca = PCA(2)
        # Transform the data
        df = pca.fit_transform(df_feat)
        # build the dataframe
        input_dataset['Cluster_Number'] = output['Cluster_Number'] + 1
        # score_metrics = score(model_ap(output, model), output)
        flyout_dict = clustering_score(output, input_dataset, 'CLARANS Clustering')
        df_1 = pd.DataFrame(df, columns=['PCA_1', 'PCA_2'])
        # predict the labels of clusters.
        label = output['Cluster_Number']
        df_1['Cluster_Number'] = output['Cluster_Number']
        scatter_plot_clustering_more_than_3cols(output, input_dataset, file_name, var)

    # Computing the time for execution as a table
    ex_time = round((time.time() - start_time), 4)
    if ex_time > 60:
        ex_minutes = int(ex_time / 60)
        ex_sec = round(ex_time % 60, 4)
        ex_time_refined = str(ex_minutes) + ' minutes ' + str(ex_sec) + ' sec'
    else:
        ex_time_refined = str(ex_time) + ' sec'

    flyout_result['Execution Time'] = ex_time_refined
    flyout_result['Number of Clusters'] = cluster_
    flyout_result['Threshold'] = thres

    # Converting the summary into a table to be displayed in the flyout
    function_summary = pd.DataFrame.from_dict(flyout_result, orient='index', columns=['Value'])

    summary = {'Parameter Values': function_summary,
               'Final Partition': final_partition(output)}

    # Inserting the model summary dataframe to the top of the flyout
    flyout_dict_new = {**summary, **flyout_dict}

    return {"TEXT": flyout_dict_new, "CSV": input_dataset}
```
